```python
!wget -O ./data/pipeline_v1.bin https://github.com/DataTalksClub/machine-learning-zoomcamp/raw/refs/heads/master/cohorts/2025/05-deployment/pipeline_v1.bin
```

```python
import pickle

with open('./data/pipeline_v1.bin', 'rb') as f_in:
    dv,model = pickle.load(f_in)

print(model)
print(dv)
```

```python
record = {
    "lead_source": "paid_ads",
    "number_of_courses_viewed": 2,
    "annual_income": 79276.0
}

X = dv.transform([record])
```

```python
score = model.predict_proba(X)
print(score)
print(score[0][1])

```
